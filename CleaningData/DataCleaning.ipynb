1. Step: Drop the empty rows

```python
df=data.dropna(how='all')
df
```

2. Step fil in the not recorded Data for ID,GENDER and E-Mail 

```python
df["id"]=df["id"].fillna(22)
df["gender"]=df["gender"].fillna("not binary or unknow")
df["email"]=df["email"].fillna("unknown")

df
```

3. Step In the Data sets is obvious wrong Data in Age= -78 and old 

```python
df["age"]=df["age"].replace("old","unknown")
df["age"]=df["age"].replace("-78","78")
df
```
